```python
#Source for next 4 blocks of code: #Source: https://github.com/otavio-s-s/data_science/blob/master/Credit%20Card%20Fraud%20Detection.ipynb
#Looking at our data
print(df.info())
print("Missing values:", df.isnull().sum())
# Check the distribution of the target variable ('Class' column)
print("Class distribution:\n", df['Class'].value_counts())
# Visualize the class distribution
sns.countplot(x='Class', data=df)
plt.title('Class Distribution')
plt.show()
```

```python
corr = df.corr()
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr, cmap='copper', vmin=-1)

plt.tight_layout()
plt.show()
```

```python
# Create a box plot with means
fig, ax = plt.subplots(figsize=(15, 8))
box_plot = sns.boxplot(x=df['Class'], y=df['Amount'], showmeans=True, ax=ax)

# Customize x-axis ticks and labels
plt.xticks([0, 1], ['Non-Fraud', 'Fraud'])

# Set y-axis limits
ax.set_ylim(0, 500)

# Set plot title and x-axis label
ax.set_title('Amount Distribution')
ax.set_xlabel('')

# Calculate and display mean values
means = df.groupby(['Class'])['Amount'].mean()
vertical_offset = df['Amount'].mean() - 130

for xtick in box_plot.get_xticks():
    box_plot.text(xtick, means[xtick] + vertical_offset, f'Mean: {round(means[xtick], 2)}', 
                  horizontalalignment='center', fontsize=14, color='w')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
df_fraud = df[df['Class'] == 1] # Recovery of fraud data
plt.figure(figsize=(15,10))
plt.scatter(df_fraud['Time'], df_fraud['Amount']) # Display fraud amounts according to their time
plt.title('Scatter plot amount fraud')
plt.xlabel('Time')
plt.ylabel('Amount in dollar')
plt.xlim([0,175000])
plt.ylim([0,2500])
plt.show()
```

```python
#create the y dataframe
y = pd.DataFrame(df.iloc[:,30])
y.columns = ['Class']

#scale the 'time' and 'amount' columns using RobustScaler()
rbs = RobustScaler()
df_portion = df[[0,29]]
df_portion = pd.DataFrame(rbs.fit_transform(df_portion))

#remove original 'time' and 'amount' columns, and add the scaled versions back into the dataframe
df_new = df.drop(labels=[0,29,30], axis=1)
df_portion.columns = [29,30]
X = pd.concat([df_new,df_portion], axis=1)
#X.head()
```

```python
# Use sklearn to split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
#Combination of under and over sampling rebalancing method
#Documentation: https://imbalanced-learn.org/stable/combine.html
#Helpful link: https://www.kaggle.com/code/rafjaa/resampling-strategies-for-imbalanced-datasets
from imblearn.combine import SMOTEENN

smote_enn = SMOTEENN(random_state=0)
X_smoteenn, y_smoteenn = smote_enn.fit_resample(X_train, y_train)

#print(len(X_smoteenn))
#print(len(y_smoteenn))
```

```python
#Oversampling rebalancing method
#Documentation: https://imbalanced-learn.org/stable/over_sampling.html
from imblearn.over_sampling import SMOTE

smote = SMOTE()
X_smote, y_smote = smote.fit_resample(X_train, y_train)

#print(len(X_smote))
#print(len(y_smote))
```

```python
#Undersampling rebalancing method
#Documentation: https://imbalanced-learn.org/stable/under_sampling.html
from imblearn.under_sampling import EditedNearestNeighbours

enn = EditedNearestNeighbours()
X_enn, y_enn = enn.fit_resample(X_train, y_train)

#print(len(X_enn))
#print(len(y_enn))
```

```python
#function that creates a report for each machine learning model
#Source: #https://towardsdatascience.com/logistic-regression-using-python-sklearn-numpy-mnist-handwriting-recognition-matplotlib-a6b31e2b166a
def create_report(y_test, pred):
    
    print("Classification Matrix")
    print(classification_report(y_test,pred))
    print("----------------------------------------------------------------")
    print()
    cr = confusion_matrix(y_test, pred)

    plt.figure(figsize=(6,6))
    sns.heatmap(cr, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
    plt.ylabel('Actual label');
    plt.xlabel('Predicted label');
    all_sample_title = 'Accuracy Score: {0}'.format(round(accuracy_score(y_test,pred)*100,2))
    plt.title(all_sample_title, size = 11);
    
    print("Confusion Matrix and Accuracy")
    plt.show()
    print("----------------------------------------------------------------")
    print()
    
predictionsList = []
```

```python
#Logistic Regression model ----------- Combination Rebalancing Technique
#Documentation: https://scikit-learn.org/stable/auto_examples/linear_model/plot_logistic.html

from sklearn.linear_model import LogisticRegression
lr1 = LogisticRegression()

#fit logistic regression model
lr1.fit(X_smoteenn,y_smoteenn)

#test model
pred_lr_smoteenn = lr1.predict(X_test)

#show results
create_report(y_test=y_test, pred=pred_lr_smoteenn)
```

```python
#Logistic Regression model ----------- Oversampling Rebalancing Technique

from sklearn.linear_model import LogisticRegression
lr2 = LogisticRegression()

#fit logistic regression model
lr2.fit(X_smote,y_smote)

#test model
pred_lr_smote = lr2.predict(X_test)

#show results
create_report(y_test=y_test, pred=pred_lr_smote)
```

```python
#Logistic Regression model ----------- Undersampling Rebalancing Technique

from sklearn.linear_model import LogisticRegression
lr3 = LogisticRegression()

#fit logistic regression model
lr3.fit(X_enn,y_enn)

#test model
pred_lr_enn = lr3.predict(X_test)

#show results
create_report(y_test=y_test, pred=pred_lr_enn)
```

```python
#Decision Tree model ----------- Combination Rebalancing Technique
#Documentation: https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html

from sklearn.tree import DecisionTreeClassifier
dt1=DecisionTreeClassifier()

#fit decision tree model
dt1.fit(X_smoteenn,y_smoteenn)

#test model
pred_dt_smoteenn = dt1.predict(X_test)

#show results
create_report(y_test=y_test, pred=pred_dt_smoteenn)
```

```python
#Decision Tree model ----------- Oversampling Rebalancing Technique

from sklearn.tree import DecisionTreeClassifier
dt2=DecisionTreeClassifier()

#fit decision tree model
dt2.fit(X_smote,y_smote)

#test model
pred_dt_smote = dt2.predict(X_test)

#show results
create_report(y_test=y_test, pred=pred_dt_smote)
```

```python
#Decision Tree model ----------- Undersampling Rebalancing Technique

from sklearn.tree import DecisionTreeClassifier
dt3=DecisionTreeClassifier()

#fit decision tree model
dt3.fit(X_enn,y_enn)

#test model
pred_dt_enn = dt3.predict(X_test)

#show results
create_report(y_test=y_test, pred=pred_dt_enn)
```

```python
#Random Forest model ----------- Combination Rebalancing Technique
#Documentation: https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html

from sklearn.ensemble import RandomForestClassifier
rf1=RandomForestClassifier(n_estimators=10)

#fit random forest model
rf1.fit(X_smoteenn,y_smoteenn)

#test model
pred_rf_smoteenn = rf1.predict(X_test)

#show results
create_report(y_test=y_test, pred=pred_rf_smoteenn)
```

```python
#Random Forest model ----------- Oversampling Rebalancing Technique

from sklearn.ensemble import RandomForestClassifier
rf2=RandomForestClassifier(n_estimators=10)

#fit random forest model
rf2.fit(X_smote,y_smote)

#test model
pred_rf_smote = rf2.predict(X_test)

#append to predictions list
predictionsList.append(pred_rf_smote)

#show results
create_report(y_test=y_test, pred=pred_rf_smote)
```

```python
#Random Forest model ----------- Undersampling Rebalancing Technique

from sklearn.ensemble import RandomForestClassifier
rf3=RandomForestClassifier(n_estimators=10)

#fit random forest model
rf3.fit(X_enn,y_enn)

#test model
pred_rf_enn = rf3.predict(X_test)

#append to predictions list
predictionsList.append(pred_rf_enn)

#show results
create_report(y_test=y_test, pred=pred_rf_enn)
```

```python
#Neural Network model ----------- Combination Rebalancing Technique
#Keras sequential model 
model = tf.keras.models.Sequential()

# Build the Neural Network model
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(32, input_dim=X_train.shape[1], activation='relu'))
model.add(tf.keras.layers.Dense(16, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X_smoteenn, y_smoteenn, epochs=10, batch_size=32, validation_split=0.2)

# Evaluate the model on the test set
#Documentation: https://machinelearningmastery.com/how-to-calculate-precision-recall-f1-and-more-for-deep-learning-models/
#Source: https://stackoverflow.com/questions/68836551/keras-attributeerror-sequential-object-has-no-attribute-predict-classes
pred_nn_smoteenn = (model.predict(X_test) > 0.5).astype("int32")

#show results
create_report(y_test=y_test, pred=pred_nn_smoteenn)
```

```python
#Neural Network model ----------- Oversampling Rebalancing Technique
#Keras sequential model 
model = tf.keras.models.Sequential()

# Build the Neural Network model
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(32, input_dim=X_train.shape[1], activation='relu'))
model.add(tf.keras.layers.Dense(16, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X_smote, y_smote, epochs=10, batch_size=32, validation_split=0.2)

# Evaluate the model on the test set
#Documentation: https://machinelearningmastery.com/how-to-calculate-precision-recall-f1-and-more-for-deep-learning-models/
#Source: https://stackoverflow.com/questions/68836551/keras-attributeerror-sequential-object-has-no-attribute-predict-classes
pred_nn_smote = (model.predict(X_test) > 0.5).astype("int32")

#show results
create_report(y_test=y_test, pred=pred_nn_smote)
```
